## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression= LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coefficients and the intercepts
print(regression.coef_)
print(regression.intercept_)
```

```python
## on which parameters the model has been trained
regression.get_params()
```

```python
## Prediction for the test data
reg_pred= regression.predict(X_test)
```

```python
reg_pred
```

```python
reg_pred.shape
```

### Assumptions of Linear Regression

```python
plt.scatter(y_test,reg_pred)
plt.xlabel("Test truth data")
plt.ylabel("Test predicted data")
```

```python
## residuals
residuals=( y_test-reg_pred)
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
## scatter plot with predictions and residuals
## uniform distribution(doesn't have any kind of shape)
plt.scatter(reg_pred,residuals)
```

```python
## Performance Matrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print('MSE:',mean_squared_error(y_test,reg_pred))

print('MAE:',mean_absolute_error(y_test,reg_pred))

print('RMSE:',np.sqrt(mean_squared_error(y_test,reg_pred)))
```

## R square and adjusted R square

### R^2=1-SSR/SST

```python
## R square 
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

### Adjusted R^2= 1- {(1-R^2)(n-1)/(n-k-1)}

```python
#adjusted R-square
1-(1-score)*(len(y_test))/(len(y_test)-X_test.shape[1]-1)
```
